```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
df=pd.read_csv("/kaggle/input/web-page-phishing-dataset/web-page-phishing.csv")
X=df.drop('phishing',axis=1)  #Target Column:phishing (1=phishing,0=legitimate)
Y=df['phishing']
X_train,X_test,Y_train,Y_test=train_test_split(
    X,Y, test_size=0.2,random_state=42
)
rf=RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    n_jobs=-1,
    class_weight='balanced'
)
rf.fit(X_train,Y_train)
rf_pred=rf.predict(X_test)
rf_acc=accuracy_score(Y_test,rf_pred)
confusion_matrix(Y_test,rf_pred)
print("Random Forest Accuracy",rf_acc)
#print(classification_report(Y_test,rf_pred))
```

## Model Comparison Table
| Model           | Accuracy |False Negatives|
|-----------------|----------|---------------|
| Decision Tree   | 88.3%    |  1393         |
| Random Forest   | 88.9%    |  890          |

### The missing phishing sites were reduced by 36%.
## Model Comparison Insight
Two models were evaluated for phishing website detection:

- Decision Tree Classifier
- Random Forest Classifier with class-weight balancing

While the Decision Tree achieved comparable accuracy, 
the Random Forest with class-weight balancing significantly reduced 
false negatives (missed phishing websites), which is critical in 
security-sensitive applications. Therefore, Random Forest was 
selected as the preferred model.
